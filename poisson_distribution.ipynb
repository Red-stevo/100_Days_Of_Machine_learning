```python
from scipy.stats import poisson
from matplotlib import pyplot as plt

lambda_ = 5.6501
events = 70000
fig, ax = plt.subplots()

rv = poisson(lambda_)
event_list = range(events)

pmf = rv.pmf(event_list)

ax.bar(event_list, pmf, log=True)
ax.set_xlabel("Number of Events.")
ax.set_ylabel("Probability of Occurence")


plt.show()
```
