```python
from scipy.stats import norm
from matplotlib import pyplot as plt
import numpy as np
#handle distrubution for mean 0 and STD of 1.

mean, std = 0, 1

rv = norm(mean, std)

# get value between 4 stds below and above the mean.
values = np.linspace(mean-4*std, mean+4*std, 1000)

probs = rv.pdf(values)

# Vitualize the data.
fig, ax = plt.subplots()

ax.plot(values, probs)
ax.set_xlabel("Values (Norm values Distribution).")
ax.set_ylabel("Probabilities.")
ax.set_title("Normal Distribution Visualization.")

```
