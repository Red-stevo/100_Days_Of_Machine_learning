```python
import random
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense, Activation, Input
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
# Working on an example 1: 

features = []
targets = []

for i in range((10 * 24891)):
    # Young with side effects.
    features.append(random.randint(5, 40))

    targets.append(1)
    
    
    # Old without side effects
    features.append(random.randint(41, 100))

    targets.append(0)

for i in range((90 * 24891)):
    # Young without side effects
    features.append(random.randint(5, 40))

    targets.append(0)


    # Old with side effects
    features.append(random.randint(41, 100))

    targets.append(1)


targets = np.array(targets)
features = np.array(features)

print(f"Targets {targets}")
print(f"features {features.shape}")
```

```python
# Data preprocessing.
scaler = MinMaxScaler(feature_range=(0, 1))
features = features.reshape(-1, 1)
print(f"Shape feature : {targets.shape}")

features, targets = shuffle(features, targets)

# Scale the data.
features = scaler.fit_transform(features)

# Training test splits
x_train, x_test, y_train, y_test = train_test_split(features, targets, test_size=0.35, random_state=42)
```

```python

# Design the neural network.
model = Sequential([
    Input(shape=(1,)),
    Dense(units=8, activation="relu"),
    Dense(units=16, activation="relu"),
    Dense(units=32, activation="relu"),
    Dense(units=2, activation="softmax")
])
```

```python
# Train the model.
model.compile(optimizer=Adam(learning_rate=0.0001), loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.fit(x=x_train, y=y_train, batch_size=100, epochs=20, shuffle=True, validation_split=0.2)
```

```python
y_pre = model.predict(x_test)
```
