```python
import os
import glob
import shutil
import random
```

```python
# paths

photos = "../../../Downloads/binary_classification/Photo/"
non_photos = "../../../Downloads/binary_classification/non-photo/"


num_photos = len(os.listdir(photos))
num_non_photo = len(os.listdir(non_photos))
```

```python
if os.path.isdir("../../../Downloads/binary_classification/dataset/train/photo") is False:
    os.mkdir("../../../Downloads/binary_classification/dataset/train/photo")
    os.mkdir("../../../Downloads/binary_classification/dataset/train/non-photo")
    os.mkdir("../../../Downloads/binary_classification/dataset/test/photo")
    os.mkdir("../../../Downloads/binary_classification/dataset/test/non-photo")
    os.mkdir("../../../Downloads/binary_classification/dataset/validate/photo")
    os.mkdir("../../../Downloads/binary_classification/dataset/validate/non-photo")

    for photo in random.sample(glob.glob(photos+"*"), int(num_photos * 0.7)):
        shutil.copy2(photo, "../../../Downloads/binary_classification/dataset/train/photo")
        
    for photo in random.sample(glob.glob(photos+"*"), int(num_photos * 0.2)):
        shutil.copy2(photo, "../../../Downloads/binary_classification/dataset/test/photo")
    
    for photo in random.sample(glob.glob(photos+"*"), int(num_photos * 0.1)):
        shutil.copy2(photo, "../../../Downloads/binary_classification/dataset/validate/photo")


    for photo in random.sample(glob.glob(non_photos+"*"), int(num_non_photo * 0.7)):
        shutil.copy2(photo, "../../../Downloads/binary_classification/dataset/train/non-photo")
        
    for photo in random.sample(glob.glob(non_photos+"*"), int(num_non_photo * 0.2)):
        shutil.copy2(photo, "../../../Downloads/binary_classification/dataset/test/non-photo")
    
    for photo in random.sample(glob.glob(non_photos+"*"), int(num_non_photo * 0.1)):
        shutil.copy2(photo, "../../../Downloads/binary_classification/dataset/validate/non-photo")

    
        
```
