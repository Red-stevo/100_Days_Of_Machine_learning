```python
import pandas as pd

data = pd.read_csv("./files/imdb_top_1000.xls")

data.head()

data.rename(columns={"Released_Year":"Year_of_Release", "Meta_score":"score", "Series_Title":"title"}, inplace=True)
```

```python
# Convert the Runtime column to int.

data.Runtime.fillna('60 min', inplace=True)
data.Runtime = data.Runtime.map(lambda d : int(d.split(" ")[0]))
```

```python
# Convert the Year_of_release column to int.
data = data.dropna()
data.Year_of_Release = pd.to_numeric(data.Year_of_Release, errors="coerce")
data.query('Year_of_Release > 2000 and Runtime > 150').sort_values(by="Year_of_Release", ascending=False)
```
