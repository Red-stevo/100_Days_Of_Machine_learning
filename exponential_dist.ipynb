```python
from scipy.stats import expon
from matplotlib import pyplot as plt
import numpy as np

lambda_ = 2.337711
scale = 1/lambda_

rv = expon(scale=scale)

values = np.linspace(1, 20, 10000)

probs = rv.pdf(values)

fig, ax = plt.subplots()

ax.set_title("Exponential Distribution Visualization.")
ax.set_xlabel("Trails")
ax.set_ylabel("Probabilities.")

ax.plot(values, probs)
```
