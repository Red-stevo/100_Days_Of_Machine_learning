```python
from scipy.stats import binom
from matplotlib import pyplot as plt

trials = 10
prob_s = 0.637

rv = binom(trials, prob_s)

trial_values = range(trials + 1)

out_comes = rv.pmf(trial_values)

fig, ax = plt.subplots()

ax.bar(trial_values, out_comes)
ax.set_title("Binomial Distribution PMF p = 0.637.")
ax.set_xlabel("Trial Number.")
ax.set_ylabel("Probabilities of Success.")
plt.style.use("dark_background")

plt.show()
```
