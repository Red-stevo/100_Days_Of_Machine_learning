```python
from scipy.stats import bernoulli as bern
from matplotlib import pyplot as plt

prob_success = 0.7
dist = bern(prob_success)
values = [1, 0]
pmf = dist.pmf(values)


print(pmf)
```

```python
plt.bar(values, pmf)
plt.title("Bernoulli Distribution (P = 0.7).")
plt.xlabel("Trial.")
plt.ylabel("Probabilities.")

plt.show()
```
